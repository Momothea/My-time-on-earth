```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sb
import json
import os
```

```python
path1 = r'C:/Users/Moise/Documents/Data visualization/My time on earth/Data/takeout-20210927T085818Z-001/Takeout/YouTube et YouTube Music/historique/watch-history.json'
with open(path1,  encoding='utf8') as json_file:
    data = json.load(json_file)
    df = pd.DataFrame(data)
```

```python
df['newtime'] = df['time'].map(pd.to_datetime)
```

```python
def get_month(df): 
    return df.month
```

```python
df['month'] = df['newtime'].map(get_month)
```

```python
df.dropna(subset = ["subtitles"], inplace=True)
```

```python
df.head()
```

```python
def get_year(df): 
    return df.year
```

```python
df['year'] = df['newtime'].map(get_year)
```

```python
def get_dom(dt):
    return dt.day
```

```python
df['day'] = df['newtime'].map(get_dom)
```

```python
def get_hour(dt):
    return dt.hour
```

```python
df['hour'] = df['newtime'].map(get_hour)
```

```python
#Adding channels
def get_channel(dt):
    #print(next(iter((dt[0]).values())))
    d = dt[0]
    return(d['name'])
```

```python
df['channel'] = df['subtitles'].map(get_channel)
```

```python
df.head()
```

```python
def get_genre(dt):
    if(dt in ['Jake Tran']):
        return "Business"
    elif(dt in ['Skill Up', 'AngryJoeShow', 'IGN', 'The Act Man', 'YongYea', 'Girlfriend Reviews', 'Upper Echelon Gamers', 'The Completionist', 'JeanBaptisteShow', 'Cleanprincegaming', 'Darksiders', 'World of Warcraft', 'The Game Theorists', 'Smosh Games', 'theRadBrad','xLetalis','Xbox', 'The Hill','RABtoons']):
        return "video game"
    elif(dt in ['Linus Tech Tips', 'Dave2D', 'Marques Brownlee', 'iJustine', 'Unbox Therapy', 'Android Authority','TechLinked','ShortCircuit','Mac Address','Techquickie','Pocketnow','Rene Ritchie','Predator Gaming']):
        return "Tech"
    elif(dt in ['Filmento', 'The Film Theorists', 'Wisecrack','Screen Junkies', 'Nerdstalgic', 'Como Deberia Haber Terminado', 'ElSmosh', 'How It Should Have Ended', 'Friendly Space Ninja', 'Screen Rant', 'Flashback FM', 'The Take','The Normies','Screen Rant','YaBoyRoshi','TV Promos','Warner Bros. Entertainment','South Park Studios','Lindsay Ellis','20th Century Studios']):
        return "Films and TV"
    elif(dt in ['freeCodeCamp.org', 'NVIDIA', 'MIT OpenCourseWare', 'Neso Academy']):
        return "Science / Programming"
    elif(dt in ['Carefree Wandering']):
        return "Philosophy"
    elif(dt in ['Skelotim', 'James Charles']):
        return "Makeup"
    elif(dt in ['Star Wars Theory', 'Star Wars Reading Club','Thor Skywalker']):
        return "Star Wars"
    elif(dt in ['ComicsExplained', 'The Imaginary Axis']):
        return "Comics"
    elif(dt in ['Taylor Davis']):
        return "Music"
    elif(dt in ['FRANCE 24', 'CNN', 'Fox News', 'The Hill']):
        return "News"
    elif(dt in ['Crunchyroll Collection', 'Naruto Explained','Plot Armor','Netflix Anime','Anime Central', 'NewWorldReview']):
        return "Manga / Anime"
    elif(dt in ['Aba & Preach']):
        return "Politics"
    else:
        return "Other"
    
    
            
```

```python
df['genre'] = df['channel'].map(get_genre)
```
